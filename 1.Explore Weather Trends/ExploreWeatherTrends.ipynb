## Project :- Exploring Weather Trends

## Importing libraries 

```python
import pandas as pd # for dealing with data!
import matplotlib.pyplot as plt # for visualizing the data!
import numpy as np# for calculating the moving average!
```

```python
globaltmp = pd.read_csv('/home/jovyan/demo/global_data.csv') # importing 'global tempreature data'
citytmp = pd.read_csv('/home/jovyan/demo/city_data.csv') # importing 'city tempreature data' which is a data for one city over multiple year.
```

## Doing 'Moving Average' for the Global data and Local data.
Moving averages are used to smooth out data and to make it easier to observe long term trends and not get lost in daily fluctuations.

```python
glb_mv_avg = globaltmp['avg_temp'].rolling(10).mean()
local_mv_avg = citytmp['avg_temp'].rolling(10).mean()
```

## Visualizng Data using matplotlib Library.
for the Global and Local Temperature data

```python
#Local Data is as same as Cairo
plt.plot(globaltmp['year'],glb_mv_avg,label='Global')
plt.plot(citytmp['year'],local_mv_avg,label='Jodhpur')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature (Â°C)") 
plt.title("Jodhpur City Average Temperature")
plt.show()
```
